#### More about words
* shape=> for shape of word eg capital,lowercase etc
* is_alpha==> returns boolean(true or false) if word is alphabeat
* is_stop==> returns boolean(true or false) if word is a stop word

```python
doc=nlp("Spacy is amazing tool")
```

```python
doc
```

```python
#word shape
for word in doc:
    print(word.text,word.shape)
```

```python
ex_doc=nlp("Hello hello HELLO HeLLO")
```

```python
for word in ex_doc:
    print("Token=>",word.text,"Shape ",word.shape_,word.is_alpha)
```

#### Parts of Speech Tagging

```python
ex1=nlp("He drinks a drink")
```

```python
ex1
```

```python
for word in ex1:
    print(word.text,word.pos_)
```

```python
ex2=nlp("I fish a fish")
```

```python
#Parts of Speech of Tag
for word in ex2:
    print(word.text,word.pos_,word.tag_)
```

```python
spacy.explain('VBP')
```

```python
excercise1=nlp(u"All the faith he had had had had no effect on the outcome of his life")
```

```python
for word in excercise1:
    print((word.text,word.tag_,word.pos_))
```

```python
spacy.explain('DET')
```

```python
excercise2=nlp("The man the professor the student has studies Rome.")
```

#### Syntactic Dependency
* It helps us to know the relation between tokens

```python
ex3=nlp("Sally likes Samy")
```

```python
ex3.text
```

```python
for word in ex3:
    print(word.text,word.tag_,word.pos_,word.dep_)
```

#### Visualizing Dependancy using Displacy

```python
# To display the dependences and any other visualization
from spacy import displacy
```

```python
# For Jupyter Notebooks you can set jupter=True to render it properly
displacy.render(ex3,style='dep')
```

#### Named Entity Recognization

```python
wikitext=nlp("By 2020 the telecom company orange, will relocate from Turkey")
```

```python
for word in wikitext.ents:
    print(word.text,word.label_)
```

```python
spacy.explain('GPE')
```

```python
from spacy import displacy
```

```python
displacy.render(wikitext,style='ent')
```

```python
spacy.explain('NORP')
```

```python
doc1=nlp("Facebook.ai,Explosion.AI,JCharisTech are all these companies")
```

```python
displacy.render(doc1,style='ent')
```
