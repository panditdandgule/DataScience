Abstract Extraction

**Steps:**

**1. Find the Most Important Words**

**2. Compute a significance score for sentences based on words they contain**

**3. Pick the topmost significant sentences**

**Word Importance **

Authors tend to repeat the words that are important to the theme of the text.

Word Importance = Word Frequency



**Sentence Significance**

Sentences which encapsulate more of the important words are more significant.

Significance Score = Sum(Word Importance)

Pick the top most significant sentences.

```python
def getTextWaPost(url):
    page = urllib.request.urlopen(articleURL).read().decode('utf8','ignore')
    soup = BeautifulSoup(page,'lxml')
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    str(text.encode('ascii',errors='replace')).replace("\xa0"," ")
    return str(text.encode('ascii',errors='replace')).replace("?"," ")
```

```python
text = getTextWaPost(articleURL)
```

```python
text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
from nltk.corpus import stopwords
from string import punctuation
```

```python
import nltk
nltk.download('punkt')
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
```

```python
word_sent = [word for word in word_sent if word not in _stopwords]
```

```python
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
print(dict(freq))
```

```python
from heapq import nlargest #find largest top 10(maximum frequency)
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict

ranking = defaultdict(int)
for i,sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
ranking
```

```python
sents_idx = nlargest(4, ranking, key=ranking.get)
```

```python
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```

```python
" ".join([sents[j] for j in sorted(sents_idx)])
```
