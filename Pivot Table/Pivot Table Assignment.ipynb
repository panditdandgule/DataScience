### Pivot Table Assignment

```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv('/home/pandit/DataScience/CodingNest/Projects/Pivot Table/train.csv')
```

```python
train.head()
```

```python
train.groupby('Sex')[['Survived']].mean()
```

```python
train.groupby(['Sex','Pclass'])['Survived'].aggregate('mean')
```

#### Using Pivot Table

```python
train.pivot_table('Survived',index='Sex',columns='Pclass')
```

```python
age=pd.cut(train['Age'],[0,20,30,80])
```

```python
train.pivot_table('Survived',['Sex','Age'],'Pclass')
```

```python
fare=pd.qcut(train['Fare'],3)
```

```python
train.pivot_table('Survived',['Sex',age],[fare,'Pclass'])
```
